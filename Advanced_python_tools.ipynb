```python
# Import NumPy
import numpy as np

# 1. Create a 1D NumPy array with integers from 1 to 20
array_1d = np.arange(1, 21)

# a. Calculate the sum, mean, median, and standard deviation
sum_1d = np.sum(array_1d)
mean_1d = np.mean(array_1d)
median_1d = np.median(array_1d)
std_1d = np.std(array_1d)

# b. Find the indices of elements greater than 10
indices_gt_10 = np.where(array_1d > 10)[0]

# 2. Create a 2D NumPy array of shape 4x4 with numbers from 1 to 16
array_2d = np.arange(1, 17).reshape(4, 4)

# a. Print the array
print("2D Array:")
print(array_2d)

# b. Transpose of the array
transpose_2d = array_2d.T

# c. Row-wise and column-wise sums
row_sums = np.sum(array_2d, axis=1)
column_sums = np.sum(array_2d, axis=0)

# 3. Create two 3x3 arrays filled with random integers between 1 and 20
array_a = np.random.randint(1, 21, size=(3, 3))
array_b = np.random.randint(1, 21, size=(3, 3))

# a. Element-wise operations
add_ab = np.add(array_a, array_b)
sub_ab = np.subtract(array_a, array_b)
mul_ab = np.multiply(array_a, array_b)

# b. Dot product of the two arrays
dot_ab = np.dot(array_a, array_b)

# 4. Reshape a 1D array of size 12 into a 3x4 2D array
array_12 = np.arange(1, 13)
array_3x4 = array_12.reshape(3, 4)

# Slice first two rows and last two columns
sliced_array = array_3x4[:2, -2:]

# Print all results
print("\n--- Results ---")
print("Sum:", sum_1d)
print("Mean:", mean_1d)
print("Median:", median_1d)
print("Standard Deviation:", std_1d)
print("Indices > 10:", indices_gt_10)
print("Transpose:\n", transpose_2d)
print("Row-wise sums:", row_sums)
print("Column-wise sums:", column_sums)
print("\nArray A:\n", array_a)
print("Array B:\n", array_b)
print("Addition:\n", add_ab)
print("Subtraction:\n", sub_ab)
print("Multiplication:\n", mul_ab)
print("Dot Product:\n", dot_ab)
print("\nOriginal 1D Array:", array_12)
print("Reshaped to 3x4:\n", array_3x4)
print("Sliced (first 2 rows, last 2 columns):\n", sliced_array)
```

Explanation

1. 1D Array Operations:

np.arange(1, 21) creates a 1D array with values from 1 to 20.

np.sum, np.mean, np.median, and np.std compute basic statistics.

np.where(array_1d > 10) finds indices of values greater than 10.

2. 2D Array (4x4):

np.arange(1, 17).reshape(4, 4) forms a 4x4 matrix.

.T gives the transpose.

np.sum(..., axis=1) and axis=0 compute row- and column-wise sums.

3. Random 3x3 Arrays:

np.random.randint(1, 21, size=(3, 3)) generates random integers.

np.add, np.subtract, np.multiply, and np.dot perform element-wise and matrix operations.

4. Reshape and Slice:

Reshaping a 1D array into 2D with .reshape(3, 4).

Slicing first 2 rows and last 2 columns with [:2, -2:].

```python
# Import Pandas
import pandas as pd

# 1. Create a DataFrame with given data
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
    'Salary': [45000, 54000, 50000, 62000, 47000]
}
df = pd.DataFrame(data)

# a. Print the first five rows of the DataFrame
print("First five rows of the DataFrame:")
print(df.head())

# b. Get summary statistics of 'Age' and 'Salary'
print("\nSummary statistics of 'Age' and 'Salary':")
print(df[['Age', 'Salary']].describe())

# c. Calculate average salary of employees in the 'HR' department
avg_hr_salary = df[df['Department'] == 'HR']['Salary'].mean()
print("\nAverage salary in HR department:", avg_hr_salary)

# 2. Add a new column 'Bonus' which is 10% of the salary
df['Bonus'] = df['Salary'] * 0.10
print("\nDataFrame with Bonus column:")
print(df)

# 3. Filter DataFrame to show employees aged between 25 and 30
filtered_df = df[(df['Age'] >= 25) & (df['Age'] <= 30)]
print("\nEmployees aged between 25 and 30:")
print(filtered_df)

# 4. Group data by 'Department' and calculate average salary
department_salary_avg = df.groupby('Department')['Salary'].mean()
print("\nAverage salary by Department:")
print(department_salary_avg)

# 5. Sort DataFrame by 'Salary' in ascending order and save to CSV
sorted_df = df.sort_values(by='Salary')
sorted_df.to_csv('sorted_employees.csv', index=False)
print("\nDataFrame sorted by Salary (saved to 'sorted_employees.csv'):")
print(sorted_df)
```

Explanation
DataFrame Creation:

A dictionary with employee data is converted into a Pandas DataFrame.

.head() shows the first five rows.

.describe() provides summary statistics for numerical columns.

Filtering by department (df[df['Department'] == 'HR']) and using .mean() gives the average salary in HR.

Bonus Column:

A new column Bonus is added by multiplying the Salary column by 0.10.

Filtering by Age:

Conditional filtering using df[(df['Age'] >= 25) & (df['Age'] <= 30)].

Grouping and Aggregation:

.groupby('Department')['Salary'].mean() calculates average salary per department.

Sorting and Saving:

.sort_values(by='Salary') sorts the DataFrame.

.to_csv('sorted_employees.csv') saves the sorted data.

```python
# Import Matplotlib and NumPy
import matplotlib.pyplot as plt
import numpy as np

# 1. Line plot
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]
plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.title('Simple Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid(True)
plt.show()

# 2. Bar Graph
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]
plt.figure(figsize=(8, 5))
plt.bar(students, marks, color=['red', 'green', 'blue', 'purple'])
plt.title('Marks Scored by Students')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.show()

# 3. Pie Chart
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]
explode = [0.1 if r == max(revenue) else 0 for r in revenue]
plt.figure(figsize=(8, 8))
plt.pie(revenue, labels=regions, autopct='%1.1f%%', startangle=90, explode=explode, shadow=True)
plt.title("Company's Revenue Distribution")
plt.axis('equal')
plt.show()

# 4. Histogram
random_data = np.random.randint(1, 101, size=1000)
plt.figure(figsize=(8, 5))
plt.hist(random_data, bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Random Integers (1-100)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

Explanation
Line Plot:

Plots x vs y using plt.plot().

Customizations include title, axis labels, and grid for better readability.

Bar Graph:

plt.bar() is used to show marks for each student.

Colors are customized, and labels and title are added.

Pie Chart:

plt.pie() displays percentage revenue from each region.

The explode parameter highlights the region with the highest revenue.

Histogram:

Generates 1000 random integers from 1 to 100 using numpy.

plt.hist() shows the frequency distribution in 20 bins.
